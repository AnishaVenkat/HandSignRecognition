```python
#Testing in real-time:

import cv2
import numpy as np
from keras.models import model_from_json
from cvzone.HandTrackingModule import HandDetector
import math
from tensorflow.keras.preprocessing.image import img_to_array

hand_dict=['Hello','ILoveYou','No','ThankYou','Yes']

# load json and create model
json_file = open('model_vgg.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
hand_model_vgg = model_from_json(loaded_model_json)

# load weights into new model
hand_model_vgg.load_weights("hand_model_vgg.h5")
print("Loaded model from disk")

detector=HandDetector(maxHands=1)

# start the webcam feed
cap = cv2.VideoCapture(0)

img_size=300

while True:
    # Find haar cascade to draw bounding box around face
    ret, frame = cap.read()
    hands,frame=detector.findHands(frame)
    offset=20
    if hands:
        hand=hands[0]
        x,y,w,h=hand['bbox']
        
        cropped_img=frame[y-offset:y+h+offset,x-offset:x+w+offset]
        
        imgwhite=np.zeros((300,300,3),np.uint8)*255
        
        imgcroppedshape=cropped_img.shape
        
        aspect_ratio=h/w
        
        if aspect_ratio>1:
            k=img_size/h
            wcal=math.ceil(k*w)
            img_resize=cv2.resize(cropped_img,(wcal,img_size))
            imgresizeshape=img_resize.shape
            Wgap=math.ceil((img_size-wcal)/2)
            imgwhite[:,Wgap:wcal+Wgap]=img_resize
        else:
            k=img_size/w
            hcal=math.ceil(k*h)
            img_resize=cv2.resize(cropped_img,(hcal,img_size))
            imgresizeshape=img_resize.shape
            Hgap=math.ceil((img_size-hcal)/2)
            imgwhite[:,Hgap:hcal+Hgap]=img_resize
        

        img=cv2.resize(imgwhite,(300,300))
        img_array=img_to_array(img)
        real_img=np.expand_dims(img_array,axis=0)
        hand_prediction = hand_model_vgg.predict(real_img)
        maxindex = int(np.argmax(hand_prediction))
        print(hand_dict[maxindex])
        cv2.putText(imgwhite, hand_dict[maxindex], (x+5, y+20), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)
            
        cv2.imshow("white image",imgwhite)
    
    key=cv2.waitKey(1)
```
